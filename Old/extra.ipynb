```python
def dfs(graph, start_node):
  """Performs a depth-first search of the graph, starting at the specified node.

  Args:
    graph: A NetworkX directed graph.
    start_node: The node to start the search at.

  Returns:
    A set of all of the nodes that were visited during the search.
  """

  # Create a stack to store the nodes that have not yet been visited.
  stack = [start_node]

  # Create a set to store the nodes that have been visited.
  visited = set()

  # While there are still nodes in the stack:
  while stack:
    # Get the next node from the stack.
    node = stack.pop()

    # Add the node to the set of visited nodes.
    visited.add(node)

    # Add all of the neighbors of the node to the stack.
    for neighbor in graph.neighbors(node):
      if neighbor not in visited:
        stack.append(neighbor)

  return visited

def has_path(graph, start_node, end_node):
  """Checks if there is a path from the start node to the end node in the graph.

  Args:
    graph: A NetworkX graph.
    start_node: The node to start the search at.
    end_node: The node to find a path to.

  Returns:
    True if there is a path from the start node to the end node, False otherwise.
  """

  # Create a set to store the nodes that have been visited.
  visited = set()

  # Create a stack to store the nodes that have not yet been visited.
  stack = [start_node]

  # While there are still nodes in the stack:
  while stack:
    # Get the next node from the stack.
    node = stack.pop()

    # If the node is the end node, return True.
    if node == end_node:
      return True

    # Add the node to the set of visited nodes.
    visited.add(node)

    # Add all of the neighbors of the node to the stack.
    for neighbor in graph.neighbors(node):
      if neighbor not in visited:
        stack.append(neighbor)

  # No path was found.
  return False
```
